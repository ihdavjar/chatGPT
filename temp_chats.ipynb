```python
new_lines = []
i = 0

while i<len(text_lines):
    temp_line = text_lines[i]
    
    for j in range(i+1,len(text_lines)):
        if text_lines[j][0:2].isdigit() and text_lines[j][2]=='/' and text_lines[j][5]=='/':
            break
        else:
            temp_line += '\n'+text_lines[j]
    
    i = j

    if (i==len(text_lines)-1):
        new_lines.append(temp_line)
        break

    new_lines.append(temp_line)
```

```python
text_lines = new_lines
```

```python
processed_lines = []
nuprocessed_lines = []

for line in text_lines:
    if " - " in line:
        if (line.split(" - ")[1]!=''):
            processed_lines.append(line.split(" - ")[1])
    else:
        nuprocessed_lines.append(line)
processed_lines
```

```python
processed_lines[0].split(":",1)[0]

temp_dict = {}
for line in processed_lines:

    if (len(line.split(":",1))>1):
        if line.split(":",1)[0] in temp_dict:

            temp_dict[line.split(":",1)[0]].append(line.split(":",1)[1])
            # temp_dict[line.split(":",1)[0]] += line.split(":",1)[1]
        
        else:
            temp_dict[line.split(":",1)[0]] = [line.split(":",1)[1]]
```

```python
text_out = ''
for line in processed_lines:
    if (len(line.split(":",1))>1):
        text_out += line.split(":",1)[1]+'\n'
```
