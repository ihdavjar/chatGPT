```python
with open('/media/ihdav/Files/Research/Infrrd_Research/chatGPT/Data/Whatsapp_Chats/WhatsApp Chat with Vadhiraj B21EE.txt','r',encoding='utf-8') as f:
    text = f.read()
```

```python
temp_lines = text.split('\n')
new_lines = []
for line in temp_lines:
    if (len(line.rsplit('-',1))>1):
        new_lines.append(line.rsplit('-',1)[1])

# temp_lines[0].rsplit('-',1)[1]
```

```python
dict_data = {}
for line in new_lines:
    if (len(line.rsplit(':',1))>1):
        
        if (line.rsplit(':',1)[0] in dict_data):
            dict_data[line.rsplit(':',1)[0]].append(line.rsplit(':',1)[1])
        else:
            dict_data[line.rsplit(':',1)[0]] = [line.rsplit(':',1)[1]]
```
